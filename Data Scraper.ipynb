```python
import re
import csv
from getpass import getpass
from time import sleep
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
URL = 'https://qcpi.questcdn.com/cdn/posting/?group=1950787&provider=1950787'
driver = webdriver.Chrome()
driver.get(URL)
page = driver.page_source
```

```python
soup = BeautifulSoup(page, "lxml")
results = soup.find(id="current_table")
print(results.prettify())
```

```python
driver.find_element(By.XPATH,'//*[@id="1"]/td[2]/a/b').click()
```

```python
list1=[]
x=driver.find_element(By.XPATH,'//*[@id="view-bid-posting"]/div[2]/div[2]/div/h4/span/b').text.split()
quest_no=x[2]
x=driver.find_element(By.XPATH,'//*[@id="view-bid-posting"]/div[2]/div[2]/div/b[2]').text.split('Date: ')
closing_date=x[1]
notes=driver.find_element(By.XPATH,'//*[@id="current_project"]/div/div[2]/div/table/tbody/tr[3]/td[2]').text
desc=driver.find_element(By.XPATH,'//*[@id="current_project"]/div/div[3]/div/table/tbody/tr[3]/td[2]').text
add_desc=driver.find_element(By.XPATH,'//*[@id="current_project"]/div/div[3]/div/table/tbody/tr[4]/td[2]').text
descrip=(quest_no,closing_date,notes,desc,add_desc)
list1.append(descrip)
n=4 #Number of pages to scrape
for i in range(n):
    driver.find_element(By.XPATH,'//*[@id="id_prevnext_next"]').click()
    sleep(4)
    x=driver.find_element(By.XPATH,'//*[@id="view-bid-posting"]/div[2]/div[2]/div/h4/span/b').text.split()
    quest_no=x[2]
    x=driver.find_element(By.XPATH,'//*[@id="view-bid-posting"]/div[2]/div[2]/div/b[2]').text.split('Date: ')
    closing_date=x[1]
    try:
        notes=driver.find_element(By.XPATH,'//*[@id="current_project"]/div/div[2]/div/table/tbody/tr[3]/td[2]').text
    except NoSuchElementException:
        notes=""
    try:
        desc=driver.find_element(By.XPATH,'//*[@id="current_project"]/div/div[3]/div/table/tbody/tr[3]/td[2]').text
    except NoSuchElementException:
        desc=""
    try:
        add_desc=driver.find_element(By.XPATH,'//*[@id="current_project"]/div/div[3]/div/table/tbody/tr[4]/td[2]').text
    except NoSuchElementException:
        add_desc=""
    descrip=(quest_no,closing_date,notes,desc,add_desc)
    list1.append(descrip)
```

```python
print(list1)
```

```python
with open('ScrapedData.csv', 'w', newline='', encoding='utf-8') as f:
    header = ['Quest_No', 'Closing date', 'Est. Notes', 'Description', 'Advanced Description']
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(list1)
```
